üì¶ What is an ‚Äúarray‚Äù?

In programming, an array is a way to store and organize data‚Äîthink of it like a list or table. In NumPy, arrays can have any number of dimensions:

1D: a list of values

2D: like a table (rows and columns)

3D: a stack of tables (like pages)

image.png

In NumPy, arrays are called ndarray (N-dimensional array).

Key rules:
All items must be the same data type

The array has a fixed size

Shape must be regular (no jagged edges)

### ‚úÖ Creating arrays


1. From list:

    Create a NumPy array from a Python list or from a sequence of elements

```python
my_list = [1, 2, 3, 4, 5, 6]
print("Python list:", my_list)

array = np.array(my_list)
print("NumPy array:", array)
```

```python
#Also like the list,

# Access elements like lists:
print(array[0])

#Arrays are mutable:
array[0] = 10
print(array)

#Use slicing:
b = array[:3]
print(b)
#Slices return views, not copies‚Äîchanging the slice changes the original array!
```

**Creating Arrays from Nested Lists**

Simulating small tabular datasets.

```python
nested_list = [
    [1, 2, 3],
    [4, 5, 6]
]
multi_array = np.array(nested_list)
print("Multi-dimensional array:\n", multi_array)
```

In NumPy, "axis" = dimension (e.g., rows = axis 0, columns = axis 1).

Access elements in multi-D arrays like: a[row, col]

```python
multi_array[1,2]
```

###‚úÖ Array attributes
Key attributes of NumPy arrays:
- `ndim`: number of dimensions
- `shape`: dimensions along each axis
- `size`:  total number of elements in array
- `dtype`: data type of elements

```python
print("Type:", type(multi_array))
print("Number of dimensions:", multi_array.ndim)
print("Shape:", multi_array.shape)
print("Size:", multi_array.size)
print("Data type:", multi_array.dtype)
```

**Statistical summary**

  Calculate mean, median, standard deviation of a dataset.

```python
data = np.array([1, 3, 5, 7, 9])
print('Mean:', np.mean(data))
print('Median:', np.median(data))
print('Std Deviation:', np.std(data))
```

### ‚úÖHandling Missing Data
- Using NaN for missing values:

  Calculate the mean of an array with missing values (use np.nanmean).

```python
data = np.array([8, np.nan, 10, 12])
print('Mean (skip nan):', np.nanmean(data))
```


## **Challenge: Small Data Science Task**

### **Scenario:**

You have a dataset with some values missing. Calculate the mean after cleaning the data, then standardize it (subtract mean, divide by std deviation).



```python
data = np.array([10, 15, np.nan, 25, 30])
mean_val = np.nanmean(data)
std_val = np.nanstd(data)
standardized = (data - mean_val) / std_val
print('Standardized:', standardized)
```

Note: np.nanstd computes standard deviation ignoring nan values.
