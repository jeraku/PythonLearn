## **Strings**<a name="string"></a>

A string represents sequence of characters. A string is represented in " " or ' ' or """ """.

**1.Initialization and Assignment**

````python
# Initializing strings using different quote types

single_quote_str = 'Hello Python!'
print(f"Single Quoted:{single_quote_str}")

double_quote_str = "Data Structures are fun!"
print(f"Double Quoted:{double_quote_str}")
````

````python
multi_line_str = """This is a
multi-line
string."""
print(f"Multi-line: \n{multi_line_str}")
````

````python
multi_line_str = """This is a \
multi-line \
string."""
print(f"Multi-line: \n{multi_line_str}")
````

````python
str1 = "Python is a powerful programming language.\nIt is widely used in data science and web development."
print(str1)
````

````python
str2 = r"Python is a powerful programming language.\nIt is widely used in data science and web development."
print(str2)
````

````python
# Initialize using str()
num = 123
string_num = str(num)
print(num,type(num),";",string_num,type(string_num)) # Output: "123"
````

````python
x = None
length = len(x)  # This will result in a TypeError
````

**5.String Slicing**

  - Syntax: **`object[start:stop:step]`**

  - Negative value for step can be used to access the elements from the end of the sequence


````python
samplestr = 'Python'   # getting substring
samplestr[slice(4)]
````

## List

- A sequence data type which can contain multiple items of different data types
- A list can be created using a square bracket ```[]``` or by using the ```list()``` method
- In general, lists are ordered, mutable, and can take duplicate values

**1.Initialization**

````python
# Initializing lists with different types of elements
empty_list = []                 # An empty list
numbers = [1, 2, 3, 4, 5]       # A list of integers
mixed = [1, "apple", 3.14, True]  # Mixed data types
nested = [[1, 2], [3, 4], [5, 6]] # A list of lists
````

````python
print(f"Empty List: {empty_list}")
print(f"Numbers: {numbers}")
print(f"Mixed: {mixed}")
print(f"Nested: {nested}")
````

````python
string_value = "hello"
char_list = list(string_value)
print(char_list)
````

**2.List Operations**

````python
value = 7
numbers.append(value) # Appends 7 to the end of the list 'numbers'
print(numbers)
````

````python
mixed.reverse() # Reverses the contents of the list, in-place
print(mixed)
````

````python
print(len(mixed)) # Prints the length (number of elements) of the list
````

**3. List Indexing and Slicing**

````python
second_item = mixed[2]
third_item_from_last = mixed[-2]
print(second_item, third_item_from_last)
````

````python
first_three_item = mixed[:3]
last_three_item = mixed[:-4:-1]
print(first_three_item, last_three_item)
````

````python
lstSample = [50,100,'C','team','India']   # list
````

````python
lstSample.index('team')   # to find the index of element 'team'
````

````python
lstSample.index('India')
````

Inserting elements at a specific index using `insert()`:

````python
print(mixed)
````

````python
mixed.insert(2,'New Zealand')

print(f"items = {mixed}")
````

Removing elements using `remove()` or `pop()`:

````python
mixed.remove('Australia')

print(f"After removing: items= {mixed}")

````

````python
mixed.pop(2)

print(f"After popping out 2nd element: items= {mixed}")
````

**4.List Concatenation**

````python
l1 = [1, 3, 5, 7]
print(l1, id(l1))
l1 += [2, 4, 6, 8]
print(l1, id(l1))
````

## Tuples

- A sequence data type which can contain multiple items of different data types
- A tuple can be created using parantheses ```()``` or by using the ```tuple()``` method
- In general, tuples are ordered, immutable, and can take duplicate values

**1.Initialization**

````python
account_balance = (10302, 57059, 8030, 14688, 103023, 28734, 57059, 8030, 14688) # Tuple Packing
print(account_balance, type(account_balance))
print(len(account_balance))
````

````python
account_holders = ('India', ['Stella', 'Rohit', 'Jessi'])
print(account_holders, type(account_holders))
print(len(account_holders))
````

**2.Tuple Operations**

````python
print(account_balance.index(10302))
````

* Unpacking Tuples - You can assign the elements of a tuple to multiple variables in a single assignment statement

````python
account_details = ('Virat', 32, 1234)
name, age, balance = account_details

print(name)
print(age)
print(balance)
````

**3. Tuple Indexing and Slicing**

````python
first_account_balance = account_balance[0]
last_account_balance = account_balance[-1]
print(first_account_balance, last_account_balance)
````

````python
first_four_accounts_balance = account_balance[1:5]
last_four_accounts_balance = account_balance[4:-1:]
print(first_four_accounts_balance, last_four_accounts_balance)
````

**4. Tuple Concatenation**

````python
a=(4)
type(a)
a=(4,)
type(a)
````

````python
tup = 1, 2, 3
print(tup, id(tup))
tup += (4,)
print(tup, id(tup))
````

- This proves that tuples are immutable

## Dictionary

- A sequence data type which can contain multiple items of different data types. The items are stored as key:value pairs
- A dictionary can also be created using flower bracket ```{}``` or by using the ```dict()``` method
- In general, dictionaries are ordered, mutable, and can not take duplicate values. By order, it means that the items in dictionaries are retained in the same order as they were when created

**1.Initialization**

````python
car_details = {
               "Brand": "Hyundai",
               "Model": "i20",
               2: 821695
              }
````

````python
car_details2 = dict([
    ("Brand", "Hyundai"),
    ("Model", "i20"),
    (2, 821695)
])
````

````python
print(car_details["Brand"])
print(car_details2[2])
````

**2. Dictionary Operations**

````python
car_details.get("Brand")
````

````python

print(car_details.keys())

print(car_details.values())

print(car_details.items())
````

````python
print(car_details.get('name'))

print(car_details.get('name', 0))
````

````python
 car_details.items()
````

````python
for key,val in car_details.items():
  print(key,":",val)
````

````python
value = car_details.pop('Model')
print(value)

key, value = car_details.popitem()
print(key)
print(value)
print(car_details)
````

**3. Dictionary concatenation**

````python
car_details.update({"Price": "1135687"})
print(car_details)
````

````python
car_details["milage"] = 19
print(car_details["milage"])
````

### Range:
* A range is a sequence of numbers that can be created using the `range()` function

* Commonly used in loops and is especially helpful for iterating through a range of numbers.

* Function can take one, two, or three arguments, and it returns a sequence of numbers starting from a specified start value, up to (but not including) an end value, and with a specified step value.

    * range(stop)

    * range(start, stop)
    
    * range(start, stop, step)

**1.Initialization**

````python
# Basic range with only stop (start defaults to 0, step defaults to 1)
r1 = range(5)  # 0, 1, 2, 3, 4
r1
````

````python
for i in r1:
    print(i)
````

````python
# Range with start and stop
r2 = range(2, 7) # 2, 3, 4, 5, 6
r2
for i in r2:
    print(i)
````

````python
# Range with start, stop, and step
r3 = range(1,12,2)
print(r3)
for i in r3:
    print(i)
````
